#### Sempre riguardo i casi totali, si chiede di scrivere del codice che, date due variabili contenenti i nomi di continenti, ne confronti i seguenti relativi descrittori statistici: valore massimo, media, e percentuale rispetto al numero dei casi totali nel mondo (calcolati anche sulle locazioni senza indicazione di continente);

```python
!pip install ipywidgets
```

```python
import ipywidgets as wdg
from IPython.display import display
```

```python
lista_continenti = df["continent"].dropna().unique().tolist()

tendina_continente1 = wdg.Dropdown(
    options = lista_continenti,
    description = "Continente 1:",
    disabled = False,
)

tendina_continente2 = wdg.Dropdown(
    options = lista_continenti,
    description = "Continente 2:",
    disabled = False,
)

def calcola_statistiche(continente1, continente2):
    data_continente1 = df[df["continent"] == continente1]["total_cases"]
    data_continente2 = df[df["continent"] == continente2]["total_cases"]

    max_continente1 = data_continente1.max()
    mean_continente1 = data_continente1.mean()
    min_continente1 = data_continente1.min()

    max_continente2 = data_continente2.max()
    mean_continente2 = data_continente2.mean()
    min_continente2 = data_continente2.min()  

    totale_mondiali = df["total_cases"].sum()
    total_continente1 = data_continente1.sum()
    total_continente2 = data_continente2.sum()

    percentuale_continente1 = (total_continente1 / totale_mondiali) * 100
    percentuale_continente2 = (total_continente2 / totale_mondiali) * 100

    print(f"\nContinente: {continente1}")
    print(f"Valore massimo dei casi: {max_continente1}")
    print(f"Media dei casi: {mean_continente1}")
    print(f"Minimo dei casi: {min_continente1}")
    print(f"Totale dei casi: {total_continente1}")
    print(f"Percentuale rispetto al totale mondiale: {percentuale_continente1:.2f}%\n")

    print(f"Continente: {continente2}")
    print(f"Valore massimo dei casi: {max_continente2}")
    print(f"Media dei casi: {mean_continente2}")
    print(f"Minimo dei casi: {min_continente2}")
    print(f"Totale dei casi: {total_continente2}")
    print(f"Percentuale rispetto al totale mondiale: {percentuale_continente2:.2f}%")

interactive_plot = wdg.interactive(calcola_statistiche, continente1=tendina_continente1, continente2=tendina_continente2)
display(interactive_plot)    
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
for country in ["Italy", "Germany", "France", "Spain"]:
    country_data = df_filtered_IGFS_2023_sorted[df_filtered_IGFS_2023_sorted["location"] == country]
    ax.plot(country_data["date"], country_data["hosp_patients"], label=country)

ax.legend()
ax.set_xlabel("Data")
ax.set_ylabel("Numero pazienti ospedalizzati")
ax.set_title("Pazienti ospedalizzati per Stato")
plt.show()
```

##### abbiamo 0 dati della germania in quel periodo quindi la ignoriamo completamente, per francia e spagna potremmo mantenere l'ultimo numero (ma diventarebbe una linea parallela all'asse x) oppure fare una media degli N valori precedenti, forse anche considerare di influenzare il risultato con il picco avuto in Italia da Novembre 2023 a inizio 2024. Questo però a sua volta non tiene conto delle diverse politiche operate dagli stati


```python
def fill_null_FS(df_filtered_IGFS_2023_sorted,country):
    country_data_temp = df_filtered_IGFS_2023_sorted[df_filtered_IGFS_2023_sorted["location"] == country].copy()
    
    for idx, row in country_data_temp.iterrows():
        if pd.isnull(row["hosp_patients"]):
            previous_values = country_data_temp.loc[:idx,"hosp_patients"].tail(11)[:-1]
            average_values = previous_values.mean()
            country_data_temp.loc[:idx,"hosp_patients"] = average_values
    return country_data_temp

france_filled = fill_null_FS(df_filtered_IGFS_2023_sorted, "France")
spain_filled = fill_null_FS(df_filtered_IGFS_2023_sorted, "Spain")
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(spain_filled['date'], spain_filled['hosp_patients'], label='Spain')
ax.plot(france_filled['date'], france_filled['hosp_patients'], label='France')

ax.legend()
ax.set_xlabel('Data')
ax.set_ylabel('Numero pazienti ospedalizzati')
ax.set_title('Pazienti ospedalizzati per Stato')

plt.tight_layout()
plt.show()
```
